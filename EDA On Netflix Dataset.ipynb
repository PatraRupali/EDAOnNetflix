```python
def unstack_col(df, col):
  Constraint = df[col].apply(lambda x:str(x).split(", ")).tolist()
  df_new1= pd.DataFrame(Constraint, index = df['title']).stack().reset_index()
  df_new1.drop("level_1", axis = 1, inplace = True)
  df_new1.rename({0:col}, axis = 1, inplace= True)
  return df_new1
```

```python
#unnesting the directors column, i.e- creating separate lines for each director in a movie
temp_df1=unstack_col(df,'director')
```

```python
#unnesting the cast column, i.e- creating separate lines for each cast member in a movie
temp_df2=unstack_col(df,'cast')
```

```python
#unnesting the country column, i.e- creating separate lines for each country in a movie
temp_df3=unstack_col(df,'country')
```

```python
#unnesting the listed_in column, i.e- creating separate lines for each genre in a movie
temp_df4=unstack_col(df,'listed_in')
```

```python
#compare shapes before merging
print(temp_df1.shape, temp_df2.shape, temp_df3.shape, temp_df4.shape)
```

```python
# DataFrames are merged sequentially using merge() on the common column title with an inner join
df_temp = temp_df2.merge(temp_df1, on='title', how='inner')  # directors + cast
df_temp = df_temp.merge(temp_df3, on='title', how='inner')   # + genre
df_new  = df_temp.merge(temp_df4, on='title', how='inner')   # + country
```

```python
#compare shapes after merging
print(temp_df1.shape, temp_df2.shape, temp_df3.shape, temp_df4.shape)
```

```python
# Looping through each column in the DataFrame
# For each column, count:
#   - Real NaN values using .isnull()
#   - Text string 'nan' values using comparison (== 'nan')
# This helps to identify both actual missing data and bad string representations of missing data
for col in df_new.columns:
    real_nan = df_new[col].isnull().sum()
    string_nan = (df_new[col] == 'nan').sum()
    print(f"Column: {col} | NaN: {real_nan} | 'nan' strings: {string_nan}")
```

```python
# Replacing missing string values in the cast and director columns with 'Unknown Actor' and 'Unknown Director' to maintain consistency in categorical analysis.
# For the country column, keeping missing values as NaN allows proper handling in statistical or geographical analysis.
df_new['cast'] = df_new['cast'].replace('nan', 'Unknown Actor')
df_new['director'] = df_new['director'].replace('nan', 'Unknown Director')
df_new['country'] = df_new['country'].replace('nan', np.nan)
```

```python
#date added column is imputed on the basis of release year,i.e- suppose there's a null for date_added
#when release year was 2013.So below piece of code just checks the mode of date added for release year=2013
# and imputes in place of nulls the corresponding mode
for i in df_final[df_final['date_added'].isnull()]['release_year'].unique():
  imp=df_final[df_final['release_year']==i]['date_added'].mode().values[0]
  
  df_final.loc[df_final['release_year']==i,'date_added']=df_final.loc[df_final['release_year']==i,'date_added'].fillna(imp)
```

```python
df_final.isnull().sum()
```

```python
known_directors = df_final[~df_final['country'].isnull()]['director'].unique()

for director in df_final[df_final['country'].isnull()]['director'].unique():
    if director != 'Unknown Director' and director in known_directors:
        mode_country = df_final[df_final['director'] == director]['country'].mode()
        if not mode_country.empty:
            df_final.loc[(df_final['director'] == director) & (df_final['country'].isnull()), 'country'] = mode_country[0]
```

```python
df_final.isnull().sum()
```

```python
known_casts = df_final[~df_final['country'].isnull()]['cast'].unique()

for cast in df_final[df_final['country'].isnull()]['cast'].unique():
    if cast != 'Unknown Actor' and cast in known_casts:
        mode_cast = df_final[df_final['cast'] == cast]['country'].mode()
        if not mode_cast.empty:
            df_final.loc[(df_final['cast'] == cast) & (df_final['country'].isnull()), 'country'] = mode_cast[0]

```

```python
#Filters rows without "Season" in duration, extracts the number, and inserts it as a new column duration_time at index 11.
temp_df_time=df_final[~df_final['duration'].str.contains('Season',na=False)]
temp_df_time.loc[:,'duration'] = temp_df_time['duration'].str.split().str.get(0)
duration_time=temp_df_time['duration']

df_final.insert(11,'duration_time',duration_time)
```

```python
#Titles such as Bahubali(Hindi Version),Bahubali(Tamil Version) were there. Since it's only one movie in different languages,
#presence of brackets and content between brackets is removed.
# More robust cleaning approach
df_final['title'] = df_final['title'].str.replace(r"\s*\(.*\)\s*", "", regex=True).str.strip()
```
